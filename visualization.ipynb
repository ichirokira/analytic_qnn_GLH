```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
gen_errs = {}
gen_errs["80"] = []
gen_errs["50"] = []
gen_errs["30"] = []
gen_errs["10"] = []
```

```python

m = 4
L = 2
std = 0.01
step_size = 1
for n in [4,8,12,16,20]:
    train_true_model_10 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_10/train_true_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))
    train_linear_model_10 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_10/train_linear_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))
    test_true_model_10 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_10/test_true_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))

    train_true_model_30 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_30/train_true_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))
    train_linear_model_30 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_30/train_linear_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))
    test_true_model_30 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_30/test_true_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))

    train_true_model_50 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_50/train_true_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))
    train_linear_model_50 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_50/train_linear_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))
    test_true_model_50 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_50/test_true_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))

    train_true_model_80 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_80/train_true_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))
    train_linear_model_80 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_80/train_linear_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))
    test_true_model_80 = np.load("/data/quaqnguy/GLH_Data/results/n={}/sample_80/test_true_model_cost_L{}_m{}_std{}_stepsize{}.npy".format(n, L, m, std, step_size))


    gen_errs["10"].append(np.abs((test_true_model_10[-1]-train_true_model_10[-1])))
    gen_errs["30"].append(np.abs((test_true_model_30[-1]-train_true_model_30[-1])))
    gen_errs["50"].append(np.abs((test_true_model_50[-1]-train_true_model_50[-1])))
    gen_errs["80"].append(np.abs((test_true_model_80[-1]-train_true_model_80[-1])))
```

```python
gen_errs_20 = []
for n_sample in [30,50,80]:
    gen_errs_20.append(gen_errs[str(n_sample)][-1]) 

plt.bar(np.arange(3), gen_errs_20, color=['darkorange', 'orange', 'bisque'])
plt.xticks(np.arange(3), [30,50,80])

```
